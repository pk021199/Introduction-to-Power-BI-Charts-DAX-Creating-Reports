1.  Define Power BI and What are the key components of the Power BI    ecosystem? Briefly explain:
 * Power BI Desktop
 * Power BI Service
 * Power BI Mobile
 * Power BI Gateway


   - **Power BI – Definition**
     
  Power BI is a business analytics and data visualization tool developed by Microsoft that enables users to connect to multiple data sources, transform and model data, create interactive reports and dashboards, and share insights across an organization to support data-driven decision-making.

  **Key Components of the Power BI Ecosystem**
  1. Power BI Desktop
     * A Windows-based application used for creating reports.
     * Allows users to:
       * Connect to various data sources (Excel, SQL Server, APIs, etc.)
       * Clean and transform data using Power Query
       * Create data models and calculations using DAX
       * Build interactive visualizations and reports
     * Reports created here can be published to the Power BI Service.

  2. Power BI Service
     * A cloud-based platform (PowerBI.com) for sharing and collaboration.
     * Used to:
       * Publish reports from Power BI Desktop
       * Create dashboards by pinning visuals
       * Share reports and dashboards with users
       * Schedule data refreshes
     * Supports collaboration and role-based access.
    
   3. Power BI Mobile
      * Mobile apps available for Android, iOS, and Windows devices.
      * Enables users to:
        * View and interact with dashboards and reports on the go
        * Receive data alerts
        * Access real-time insights anytime, anywhere
        * Optimized for smaller screens.

   4. Power BI Gateway
      * A bridge between on-premises data sources and the Power BI Service.
      * Allows Power BI Service to:
        * Access on-premises databases securely
        * Refresh data automatically without moving data to the cloud
        * Commonly used with SQL Server, Oracle, and other on-premises systems.

  **Summary**
   
   Power BI Desktop is used to create reports, Power BI Service is used to publish and share, Power BI Mobile allows access on mobile devices, and Power BI Gateway enables secure data connectivity between on-premises data and the cloud.

2. Compare the following Power BI visuals:
  * Pie Chart vs Donut Chart
  * Bar Chart vs Column Chart
When would you prefer one over the other? Give one example for each pair.


 - **1. Pie Chart vs Donut Chart (Definition-Based Comparison)**
   * Pie Chart:
     
     A Pie Chart is a circular visual that represents data as slices of a whole, where each slice shows the proportion of a category relative to the total.

   * Donut Chart:
     
     A Donut Chart is a variation of a pie chart with a hole in the center, allowing space to display additional information such as totals or KPIs while still showing part-to-whole relationships.

     **When to Prefer:**
     * Prefer a Pie Chart when you want a simple and quick view of proportions with very few categories.
     * Prefer a Donut Chart when you want to display totals or key values in the center along with proportions.

     **Example:**
     * Pie Chart: Percentage share of 3 product categories in total sales.
     * Donut Chart: Sales contribution by region with total sales shown in the center.
   **2. Bar Chart vs Column Chart (Definition-Based Comparison)**
    * Bar Chart:
     
      A Bar Chart displays data using horizontal bars, making it suitable for comparing values across categories, especially when category names are long.
    * Column Chart:
      
      A Column Chart displays data using vertical columns, commonly used to compare values across time periods or show trends.

      **When to Prefer:**
      * Prefer a Bar Chart when comparing many categories or when category labels are long.
      * Prefer a Column Chart when analyzing time-based data or showing changes over time.

      **Example:**
      * Bar Chart: Comparison of revenue across different product names.
      * Column Chart: Monthly sales performance over one year.

3. Explain the significance of:
  * Star schema vs Snowflake schema
  * Primary key vs Foreign key in relationships (Power BI)
    Why is cardinality important?


 - **1. Star Schema vs Snowflake Schema – Significance**

     **Star Schema:**
     A Star Schema is a data model where a central fact table is directly connected to multiple dimension tables. It has a simple structure and is easy to understand.

     **Significance:**
      * Improves query performance
      * Easier to design and maintain
      * Preferred and recommended in Power BI for faster reporting
    
     **Snowflake Schema:**
      A Snowflake Schema is a more complex model where dimension tables are normalized into multiple related tables.

     **Significance:**
      * Reduces data redundancy
      * More complex relationships
      * Slightly slower performance compared to star schema

     **Summary:**
       Power BI generally performs better with a Star Schema, while Snowflake Schema is useful when data normalization is required.


   **2. Primary Key vs Foreign Key in Relationships (Power BI)**

     **Primary Key:**
     A Primary Key is a column that uniquely identifies each record in a table (usually in a dimension table).

     **Foreign Key:**
     A Foreign Key is a column in one table that references the primary key of another table (usually in a fact table).

     **Significance in Power BI:**
      * Primary keys ensure unique values
      * Foreign keys create relationships between tables
      * Enable accurate filtering and aggregation across tables

   **3. Why Is Cardinality Important in Power BI?**
      
      Cardinality defines the relationship type between tables (One-to-One, One-to-Many, Many-to-One, Many-to-Many).

     **Importance:**
      * Controls how filters flow between tables
      * Ensures correct calculations and totals
      * Improves data model performance
      * Prevents ambiguous or incorrect relationships

     **Example:**
      In a sales model, one customer (dimension) can have many sales records (fact), so the correct cardinality is One-to-Many.

     **Conclusion**
      Using a Star Schema, correctly defining Primary and Foreign Keys, and setting proper cardinality are essential for accurate, efficient, and reliable Power BI reports.

4. Differentiate between:
   
   Calculated column vs Measure
     Also, define Row context and Filter context with simple examples.


 - **Calculated Column vs Measure**

   * Calculated Column:
     A calculated column is a new column created in a table using DAX, where the calculation is performed row by row and the result is stored in the data model.
     * Evaluated during data refresh
     * Increases model size
     * Used for row-level logic and categorization
   * Example:
     
     Profit = Sales[Revenue] - Sales[Cost] (calculated for each row)

   * Measure:
     A measure is a DAX calculation that is evaluated at runtime based on the current filters in a report and is not stored in the data model.
     * Calculated dynamically
     * Uses filter context
     * Ideal for aggregations and KPIs
   * Example:
     
     Total Sales = SUM(Sales[Revenue])

 * **Row Context**
     
     Row context refers to the current row being evaluated during a calculation.
     * Applies mainly to calculated columns
     * Each row is calculated independently
     
     Example:
     
     If a table has Quantity and Price columns, a calculated column:

     Total = Sales[Quantity] * Sales[Price]

     Here, Power BI evaluates each row separately.

 * **Filter Context**
     
     Filter context refers to the set of filters applied to a calculation from visuals, slicers, rows, columns, or report filters.
     * Applies mainly to measures
     * Changes dynamically with user interaction
     
     Example:

     A measure:
     Total Sales = SUM(Sales[Revenue])
     
     If a slicer selects Year = 2024, the measure calculates total sales only for 2024.

5. What is the difference between a report and a dashboard in Power BI?
 - **Difference Between Report and Dashboard in Power BI**
   
   **Report:**
   A report in Power BI is a multi-page, detailed collection of visuals created from a single dataset. It is mainly used for in-depth analysis and exploration of data.
     * Created in Power BI Desktop
     * Can have multiple pages
     * Supports slicers, filters, and drill-down
     * Used for detailed and interactive analysis
   
   Example:
     A sales report with separate pages for sales by region, product, and time.

   **Dashboard:**
     A dashboard in Power BI is a single-page view made up of pinned visuals from one or multiple reports and datasets. It provides a high-level summary of key metrics.
      * Created only in Power BI Service
      * Always one page
      * Focuses on KPIs and highlights
      * Used for quick monitoring

   Example:
     An executive dashboard showing total sales, profit, and growth rate.

10. Scenario:
     VitaTrack Wellness, a digital health company in FitZone, has collected data on users’
     daily habits and health vitals. The analytics team is tasked with drawing actionable
     insights from this data to improve lifestyle suggestions and prevent heart-related risks.

     **Your Task:**
     
     Using the provided dataset (includes Age, Gender, BMI, Steps, Calories, Sleep, Heart Rate, Blood Pressure, Smoking, Alcohol, Exercise, Diabetic & Heart Disease status):

     **Build a one-page Power BI dashboard that answers:**
      1. Are users maintaining a balanced lifestyle (Steps, Sleep, Calories)
      2. What lifestyle patterns (Smoking, Alcohol, BMI, etc.) indicate heart disease risk?
      3. Is there any visible relationship between Sleep and Physical Activity?
      4. How does BMI vary across Age Groups and Genders?
      5. What is the impact of smoking and alcohol on heart rate and blood pressure?
      6. Segment people based on their health activity to suggest lifestyle changes

**Power BI Dashboard: VitaTrack Wellness – Lifestyle & Heart Health Insights**
1. Balanced Lifestyle Overview
     
     Visuals:
  * Gauge Charts / KPI Cards:
     * Average Daily Steps vs WHO recommendation (≥ 10,000 steps).
     * Average Sleep Hours vs recommended 7–9 hours.
     * Average Calories Intake vs recommended 2000–2500 kcal.
     
     Insight: Quickly shows whether the majority of users are meeting healthy lifestyle thresholds.
2. Lifestyle Patterns & Heart Disease Risk
     
     Visuals:
  * Stacked Bar Chart: Heart Disease prevalence segmented by Smoking, Alcohol    Consumption, BMI categories (Normal, Overweight, Obese).
  * Heatmap: Correlation between BMI, Smoking, Alcohol and Heart Disease status.
     
     Insight: Identifies high-risk groups (e.g., smokers with high BMI and alcohol intake show elevated heart disease incidence).
3. Sleep vs Physical Activity
     
     Visuals:
  * Scatter Plot: Sleep Hours vs Daily Steps, color-coded by Exercise Hours per Week.
  * Trend Line: Shows whether more active individuals also sleep better.
     
     Insight: Reveals if poor sleep is linked to low activity or if high activity correlates with better rest.
4. BMI Across Age Groups & Gender
     
     Visuals:
  * Box & Whisker Plot: BMI distribution across Age Groups (18–30, 31–45, 46–60, 61+).
  * Clustered Column Chart: BMI averages by Gender.
     
     Insight: Highlights age ranges and genders with higher obesity prevalence.
5. Impact of Smoking & Alcohol on Heart Rate & Blood Pressure
     
  * Dual Axis Line Chart: Compare Heart Rate and Blood Pressure for smokers vs non-smokers.
  * Clustered Bar Chart: Alcohol consumption levels vs average Heart Rate and Systolic/Diastolic BP.
     
     Insight: Shows how lifestyle choices directly affect cardiovascular vitals.

6. Segmentation for Lifestyle Suggestions
     
     Visuals:
  * Clustered Segmentation (K-Means or Power BI Clustering):
     * Group 1: Active, balanced lifestyle (high steps, good sleep, normal BMI).
     * Group 2: Sedentary, high risk (low steps, poor sleep, high BMI, smoker).
     * Group 3: Moderate lifestyle (average steps, borderline BMI, occasional alcohol).
     * Group 4: High activity but poor sleep (risk of fatigue/stress).

     Insight: Enables personalized recommendations (e.g., Group 2 → increase activity, reduce smoking/alcohol; Group 4 → improve sleep hygiene).

**Suggested Layout (One-Page Dashboard)**
  * Top Row (KPIs): Steps, Sleep, Calories gauges.
  * Middle Row (Risk Analysis): Lifestyle patterns vs heart disease, Sleep vs Activity scatter.
  * Bottom Row (Demographics & Segmentation): BMI by age/gender, Smoking/Alcohol impact, Lifestyle clusters.


 This design ensures executives and health coaches can instantly see where interventions are needed, while users get personalized lifestyle improvement suggestions.
